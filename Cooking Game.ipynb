# Here's a simple game I created for my year-1 final project which required utilizing Object-Oriented Programming and State Machines. This was my first experience writing such an extensive code. Minor improvements can be made to make the class Dishes to inherit from class Food or just combine the two. However, this sparked my interest to further improve and develop on my coding capabilities

```python
from libdw import sm
import random
import time

class Player:
    def __init__(self, name, level, hunger):
        self.name = name
        self.hunger = hunger
        self.level = level
        
    def intro(self):
        if self.level == 1:
            return "Hi " + self.name + "! You can cook some basic stuff I guess... Let's prepare something together since you're " + self.is_hungry()
        if self.level == 2:
            return "Hi " + self.name + "! You're an alright cook. Let's prepare something together since you're " + self.is_hungry()
        else:
            return "Hi " + self.name +  "! You're a Masterchef! Let's prepare something together since you're " + self.is_hungry()
    
    def is_hungry(self):
        if self.hunger == 0:
            return "so full"
        elif self.hunger == 1:
            return "in need of a snack"
        elif self.hunger == 2:
            return "a little hungry"
        else:
            return "SO HUNGRY"
    
    def eat(self):
        self.hunger = 0

class Food:
    def __init__(self, dish, fill, ingredients):
        self.dish = dish
        self.fill = fill
        self.ingredients = ingredients
        
    def info(self):
        return "You will be making " + self.dish

class Dishes:
    def __init__(self, fill):
        self.fill = fill
        
    def food_creator(self):
        fill_1 = {'Sliced Fruits':['apple', 'banana', 'strawberry', 'dragonfruit'], 'Boiled Eggs':['egg', 'water'], 'Cereal': ['milk', 'honeystars', 'cornflakes'], 'Toast':['bread', 'butter']}
        fill_2 = {'Cheeseburgers':['hamburger bun','meat patty','cheese'], 'Hotdogs':['hotdog bun', 'hotdog'], 'Sandwiches':['bread','cheese','lettuce','ketchup'], 'Prata':['dough','curry','egg']}
        fill_3 = {'Chicken Rice':['chicken','rice','chilli'], 'Fish and Chips':['cod fish', 'seabass', 'fries','tartar sauce'], 'Wanton Mee(HALAL)':['noodles','chicken char siew','wanton'], 'Fried Rice':['rice','egg','fried shallots'], 'Briyani':['rice','ghee','meat','spices']}
        if self.fill == 1:
            ingredients = list(fill_1.values())
            dish = random.choice(list(fill_1.keys()))
            f = Food(dish, self.fill, ingredients)
        elif self.fill == 2:
            ingredients = list(fill_2.values())
            dish = random.choice(list(fill_2.keys()))
            f = Food(dish, self.fill, ingredients)
        else:
            ingredients = list(fill_3.values())
            dish = random.choice(list(fill_3.keys()))
            f = Food(dish, self.fill, ingredients)
        return f
    
    def display_info(self, i):
        f = self.food_creator()
        if i == 1:
            print(f.info())
        elif i == 2:
            return f.ingredients
        

class Game(sm.SM):
    def __init__(self):
        self.start_state = ["play","nil","raw"]
        self.count = 0
        self.clicks = 0
        self.done = 0
        self.P = None
        self.dish = None
    
    def press(self, i):
        self.clicks += i
    
    def set_diff(self):
        if self.P.level == 1:
            self.count = 8
        elif self.P.level == 2:
            self.count = 10
        else:
            self.count = 7
    
    def get_next_values(self, state, inp):
        if state[0] == "play":
            if state[1] == "nil":
                if inp == 1:
                    next_state = ["play", "basic", "raw"]
                    info_str = "You will be playing the basic level"
                elif inp == 2:
                    next_state = ["play", "mod", "raw"]
                    info_str = "You will be playing the moderate level"
                elif inp == 3:
                    next_state = ["play", "hard", "raw"]
                    info_str = "You will be playing the hard level"
                return next_state, info_str
                
            if state[1] == "basic":
                if state[2] == "raw":
                    if inp == 1:
                        next_state = ["play", "basic", "cooking"]
                        self.press(inp)
                        info_str = "You have started cooking! Press 1-9 to keep cooking, press 0 to find out how long till you're done\t"
                    else:
                        next_state = state
                        info_str = "That's not right :( Press 1 start cooking\t"
                    return next_state, info_str
                
                elif state[2] == "cooking":
                    if inp == 0:
                        next_state = state
                        info_str = "You have " + str(self.count - self.clicks) + " more clicks to go! Press 1-9 to keep cooking\t"
                    elif inp <= (self.count - self.clicks):
                        self.press(inp)
                        while self.clicks < self.count:
                            info_str = "Keep going! " + str(self.count - self.clicks) + " more to go! Press 1-9 to keep cooking\t"
                            next_state = ["play", "basic", "cooking"]
                            return next_state, info_str
                        next_state = ["play", "basic", "cooked"]
                        info_str = "Cooked perfectly! Press 1 to eat\t"
                    elif inp > (self.count - self.clicks):
                        next_state = ["play", "basic", "overcooked"]
                        info_str = "You burnt it! Press 1 to eat your burnt food (lol) or press 0 to restart\t"
                    else:
                        next_state = state
                        info_str = "That's not right :( Press 1-5 to keep cooking\t"
                    return next_state, info_str
                    
                elif state[2] == "cooked":
                    if inp == 1:
                        next_state = ["eaten", "basic", "end"]
                        info_str = "You ate it! Hope it was good\t"
                        self.P.eat()
                        self.end()
                    else:
                        next_state = state
                        info_str = "That's not right :( Press 1 to eat it before it becomes stale :O\t"
                    return next_state, info_str
                
                elif state[2] == "overcooked":
                    if inp == 0:
                        self.clicks = 0
                        next_state = ["play", "basic", "raw"]
                        info_str = "Restarting"
                    elif inp == 1:
                        next_state = ["eaten", "basic", "end"]
                        info_str = "You ate it but hope you know eating burnt food isnt good for you...."
                        self.P.eat()
                        self.end()
                    else:
                        next_state = state 
                        info_str = "That's not right :( Press 1 to eat your burnt food (lol) or press 0 to restart\t"
                    return next_state, info_str
                    
            elif state[1] == "mod":
                if state[2] == "raw":
                    if inp == 1:
                        next_state = ["play", "mod", "cooking"]
                        self.press(inp)
                        info_str = "You have started cooking! Press between 1-5 to continue cooking\t"
                    else:
                        next_state = state
                        info_str = "That's not right! Press 1 to start cooking!\t"
                    return next_state, info_str
                    
                elif state[2] == "cooking":
                    if inp <= (self.count - self.clicks):
                        if inp != 0:
                            self.press(inp)
                            while self.clicks <= self.count:
                                next_state = ["play", "mod", "cooking"]
                                self.sounds()
                                info_str = "Press between 1-5 to continue cooking, press 0 if you think it's cooked!\t"
                                return next_state, info_str
                            next_state = ["play", "mod", "cooked"]
                            info_str = "Perfectly cooked! Press 1 to eat it\t"
                            return next_state, info_str
                        elif inp == 0:
                            if self.clicks == self.count:
                                next_state = ["play", "mod", "cooked"]
                                info_str = "Perfectly cooked! Press 1 to eat it\t"
                            else:
                                next_state = ["play", "mod", "cooking"]
                                info_str = "Oh no it's still raw :( Press between 1-5 to continue cooking\t"
                            return next_state, info_str
                    else:
                        next_state = ["play", "mod", "overcooked"]
                        info_str = "You burnt it! Press 1 to eat your burnt food (lol) or press 0 to restart\t"
                        return next_state, info_str

                elif state[2] == "cooked":
                    if inp == 1:
                        next_state = ["eaten", "mod", "end"]
                        info_str = "You ate it! Hope you are full now :)"
                        self.P.eat()
                        self.end()
                        return next_state, info_str
                    
                elif state[2] == "overcooked":
                    if inp == 1:
                        next_state = ["eaten", "mod", "end"]
                        info_str = "You ate it but hope you know eating burnt food isnt good for you...."
                        self.P.eat()
                        self.end()
                        return next_state, info_str
                    elif inp == 0:
                        self.clicks = 0
                        next_state = ["play", "mod", "raw"]
                        info_str = "Restarting"
                        return next_state, info_str
                    
            elif state[1] == "hard":
                if state[2] == "raw":
                    if inp == 1:
                        self.press(inp)
                        next_state = ["play", "hard", "cooking"]
                        info_str = "You have started cooking! Brace yourselves for some questions on the dish..."
                        return next_state, info_str
                    
                elif state[2] == "cooking":
                    if inp == 1:
                        while self.clicks < self.count:
                            self.press(inp)
                            next_state = ["play", "hard", "cooking"]
                            info_str = self.hard_question()
                        next_state = ["play", "hard", "cooked"]
                        info_str = "You answered everything right... I am impressed. Your dish is cooked! Press 1 to eat it\t"
                    else:
                        next_state = ["play", "hard", "raw"]
                        info_str = "You're wrong! You sure you're a level 3 chef...? Back to the beginning"
                        self.clicks = 0
                    return next_state, info_str
                
                elif state[2] == "cooked":
                    if inp == 1:
                        next_state = ["eaten", "hard", "end"]
                        info_str = "You ate your marvellous creation :)"
                        self.P.eat()
                        self.end()
                        return next_state, info_str
                    else:
                        next_state = state
                        info_str = "Hello? Press 1 to eat!\t"
                        return next_state, info_str
                    
    def sounds(self):
        action = ["Preparing the ingredients....", "Heating up the pan....", "Adding the oil....", "Cooking the ingredients....", "Tasting the seasoning...." ]
        season = ["Adding the salt....", "Adding the pepper....", "Adding some msg....", "Adding some sugar...."]
        if self.clicks < 5:
            output = "\n" + action[self.clicks] + "\n"
            print(output)
        else:    
            if self.clicks % 2 == 0:
                output = "\n" + action[4] + "\n"
                print(output)
            else:
                output = "\n" + random.choice(season) + "\n"
                print(output)
        
    def end(self):
        self.done += 1

    def play(self): 
        print("Whats your name?")
        name = input("Name:\t")

        print("How hungry are you from 1-3?")
        hunger = (input("Hunger 1-3:\t")).strip()
        while self.done == 0:
            try:
                hunger = int(hunger)
                if hunger == 1:
                    break
                elif hunger == 2:
                    break
                elif hunger == 3:
                    break
                else:
                    print("Press 1-3 only please")
                    hunger = (input("Hunger 1-3:   ")).strip()
                        
            except ValueError:
                print("Press 1-3 only please")
                hunger = (input("Hunger 1-3:    ")).strip()
        
        print("How good is your cooking from 1-3?")
        level = (input("Level 1-3:\t")).strip()
        while self.done == 0:
            try:
                level = int(level)
                if level == 1:
                    break
                elif level == 2:
                    break
                elif level == 3:
                    break
                else:
                    print("Press 1-3 only please")
                    level = (input("Level 1-3:    ")).strip()
                    
            except ValueError:
                print("Press 1-3 only please")
                level = (input("Level 1-3:    ")).strip()
        
        self.P = Player(name, level, hunger)
        self.end()
        key = input("Press 1 to start the game. Press 2 to find out what dish you're cooking!\n>>>")
        
        while self.done == 1:
            try:
                key = int(key)
            except ValueError:
                key = input("Press 1 or 2 only please\n>>>")
            if key == 1:
                self.start()
                self.end()
            elif key == 2:
                self.get_dish()
                key = input("Press 1 if you want to cook this. Press 2 if you want to change it\n>>>")
            else:
                key = input("Press 1 or 2 only please\n>>>")

        self.set_diff()
        while self.done == 2:
            while self.state[0] == "play":
                if self.state[1] == "nil":
                    if self.P.level == 1:
                        print(self.step(1))
                    elif self.P.level == 2:
                        print(self.step(2))
                    elif self.P.level == 3:
                        print(self.step(3))
                else:
                    if self.clicks == 0:
                        key = input("Press 1 to start cooking!\n>>>")
                        try:
                            key = int(key)
                            print(self.step(key))
                        except ValueError:
                            print("What are you doing? Press 1 to start cooking!\n>>>")
                    else:
                        if self.state[1] != "hard":
                            key = input(">>>")
                            try:
                                key = int(key)
                                print(self.step(key))
                            except ValueError:
                                print("What are you doing? Press 1 to start cooking!\n>>>")
                        else:
                            if self.state[2] == "cooking":
                                key = self.hard_question()
                                print(self.step(key))
                            elif self.state[2] == "cooked":
                                key = input(">>>")
                                try:
                                    key = int(key)
                                    print(self.step(key))
                                except ValueError:
                                    print("What are you doing? Press 1 to eat it!\n>>>")
        time.sleep(2)
        print("\n-----GAME OVER-----\n")
        time.sleep(2)
        print("Thank you for playing :)")
        
    def get_dish(self):
        self.dish = Dishes(self.P.hunger)
        self.dish.display_info(1)
        
    def hard_question(self):
        ingredients = self.dish.display_info(2)
        some_ingredients = ['chicken', 'apple', 'dragonfruit', 'egg','honeystars','bread','ghee','cod fish', 'cheese', 'chicken char siew', 'curry']
        choice = random.choice(some_ingredients)
        
        question = "\nIs " + choice + " an ingredient? Yes or No \n"
        print(question)
        answer = (input("Yes/No:\t")).strip()
        
        while self.state[1] == "hard":
            if choice in ingredients:
                if answer.isalpha():
                    if answer.lower() == "yes":
                        return 1
                    elif answer.islower() == "no":
                        return 0
                    else:
                        print("Answer Yes or No please\n")
                        answer = (input("Yes/No:\t")).strip()
                else:
                    print("Answer Yes or No please\n")
                    answer = (input("Yes/No:\t")).strip()
            elif choice not in ingredients:
                if answer.isalpha():
                    if answer.lower() == "no" :
                        return 1
                    elif answer.lower() == "yes":
                        return 0
                    else:
                        print("Answer Yes or No please\n")
                        answer = (input("Yes/No:\t")).strip()
                else:
                    print("Answer Yes or No please\n")
                    answer = (input("Yes/No:\t")).strip()
    
g = Game()
g.play()
```
